### 13-) What is the average consuming price for the USA citizens ?

```python
usa_cons_rank = pd.read_sql("""
SELECT AVG("Price")
FROM Orders o 
LEFT JOIN orderdetails od on o."OrderID" = od."OrderID" 
LEFT JOIN products p on od."ProductID" = p."ProductID" 
LEFT JOIN customers cu on o."CustomerID" = cu."CustomerID" where cu."Country" = 'USA';
""", conn)
usa_cons_rank.head()
```

### 14-) Please sort the employee names with their total sales ?

```python
emp_rank = pd.read_sql("""
SELECT em."FirstName", SUM(p."Price") as total_sales
FROM orders o 
LEFT JOIN orderdetails od on o."OrderID" = od."OrderID" 
LEFT JOIN products p on od."ProductID" = p."ProductID" 
LEFT JOIN employees em on o."EmployeeID" = em."EmployeeID" 
GROUP BY em."FirstName"
ORDER BY total_sales DESC;
""", conn)
emp_rank
```

### 15-) Please provide the average category product price in Products Table ?

```python
cat_rank = pd.read_sql("""
SELECT "CategoryName", AVG(p."Price") as avg_price
FROM products as p
LEFT JOIN categories ca on p."CategoryID" = ca."CategoryID"
GROUP BY "CategoryName"
""", conn)
cat_rank
```

### 16-) Please provide average category price orders from Germany ?

```python
cat_ge = pd.read_sql("""
SELECT ca."CategoryName", AVG(p."Price") 
FROM orders o 
LEFT JOIN orderdetails od on o."OrderID" = od."OrderID" 
LEFT JOIN products p on od."ProductID" = p."ProductID" 
LEFT JOIN categories ca on p."CategoryID" = ca."CategoryID" 
LEFT JOIN customers cu on o."CustomerID" = cu."CustomerID" 
WHERE cu."Country" = 'Germany' 
GROUP BY "CategoryName"
""", conn)
cat_ge
```

### 17-) What is the average order price from Germany or USA based on their categories?

```python
cat_ge_usa = pd.read_sql("""
SELECT ca."CategoryName", AVG(p."Price") 
FROM orders o 
LEFT JOIN orderdetails od on o."OrderID" = od."OrderID" 
LEFT JOIN products p on od."ProductID" = p."ProductID" 
LEFT JOIN categories ca on p."CategoryID" = ca."CategoryID" 
LEFT JOIN customers cu on o."CustomerID" = cu."CustomerID" 
WHERE cu."Country" = 'Germany' OR cu."Country" = 'USA'
GROUP BY "CategoryName"
""", conn)
cat_ge_usa
```

### 18-) What is the average price for June, July and August ?

```python
report = pd.read_sql("""
SELECT AVG(p."Price")
FROM products as p
LEFT JOIN orderdetails od on p."ProductID" = od."ProductID"
LEFT JOIN orders o on  od."OrderID" = o."OrderID"
WHERE "OrderDate" LIKE '%07%' or "OrderDate" LIKE '%08%' or "OrderDate" LIKE '%09%'
""", conn)
report
```

### 19-) Find the maximum quantities of customer orders for 1997 and list them in descending order with the names of the customers ?

```python
report_1997 = pd.read_sql("""
SELECT "CustomerName", MAX(p."Price") as max_price
FROM products as p
LEFT JOIN orderdetails od on p."ProductID" = od."ProductID"
LEFT JOIN orders o on  od."OrderID" = o."OrderID"
LEFT JOIN customers on customers."CustomerID" = o."CustomerID" 
WHERE "OrderDate" LIKE '%1997%'
GROUP BY "CustomerName"
ORDER BY max_price DESC
""", conn)
report_1997
```
